## Loop

```python
def main():
    number = get_number()
    meow(number)

def get_number():
    while True:
        n = int(input("give a number "))
        if n > 0:
            break
    return n

def meow(n):
    for _ in range(n):
        print("meow")

main()
```

```python
def main():
    print_square(get_number())

def get_number():
    while True:
        n = int(input("give a number "))
        if n > 0:
            break
    return n

def print_square(n):
    for i in range(n):
        for j in range(n):
            print("#", end=" ")
        print()

main()
```

## Error handling

```python
try:
    x = int(input("what's X? "))
    print(f"x is {x}")

except ValueError:
    print("That's not an integer!")
```

```python
def main():
    x = get_number()
    print(f"x is {x}")

def get_number():
    while True:
        try:
            n = int(input("what's X? "))
        except ValueError:
            print("That's not an integer!") # can also use Pass here
        else:
            return n # can be break or

main()
```
